```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

## Vulture

```python
raw = pd.read_csv('./vultures.csv', header=None)

X = []
Y = []

for i in range(len(raw)):
    
    row = np.array(raw.loc[i][~raw.loc[i].isnull()])
    data = row[0:-1].astype(float)
    label = row[-1]
    
    x_axis = data[0::3]
    y_axis = data[1::3]
    z_axis = data[2::3]
    
    X.append(np.stack((x_axis, y_axis, z_axis)))
    Y.append(label)

Y = np.array(list(enumerate(Y)))
```

```python
plot_data = {}

for cls in np.unique(Y[:, 1]):
    temp = Y[Y[:, 1] == cls]
    idx = int(random.choice(temp)[0])
    data_choice = X[idx]
    plot_data[cls] = data_choice.transpose((1, 0))
```

```python
f, axarr = plt.subplots(2, 3, sharex=False, sharey=False, figsize=(10, 5))

i = 0
for beh in plot_data.keys():
    ax = axarr.reshape((-1))[i]
    ax.plot(range(len(plot_data[beh])), plot_data[beh][:, 0], label='x-axis')
    ax.plot(range(len(plot_data[beh])), plot_data[beh][:, 1], label='y-axis')
    ax.plot(range(len(plot_data[beh])), plot_data[beh][:, 2], label='z-axis')
    ax.set_xlim([0, 82])
    ax.set_ylim([-30, 30])
    ax.xaxis.set_major_locator(plt.MaxNLocator(3))
    ax.yaxis.set_major_locator(plt.MaxNLocator(3))
    ax.set_title(beh)
    i += 1
ax.legend(loc=10, bbox_to_anchor=(1.3, 0.2))

plt.tight_layout()

plt.show()
```

## Storks

```python
raw = pd.read_csv('./storks_obs_train.csv', header=None)

X = []
Y = []

for i in range(len(raw)):
    
    row = np.array(raw.loc[i][~raw.loc[i].isnull()])
    data = row[0:-1].astype(float)
    label = row[-1]
    
    x_axis = data[0::3]
    y_axis = data[1::3]
    z_axis = data[2::3]
    
    X.append(np.stack((x_axis, y_axis, z_axis)))
    Y.append(label)

Y = np.array(list(enumerate(Y)))
```

```python
plot_data = {}

for cls in np.unique(Y[:, 1]):
    temp = Y[Y[:, 1] == cls]
    idx = int(random.choice(temp)[0])
    data_choice = X[idx]
    plot_data[cls] = data_choice.transpose((1, 0))
```

```python
f = plt.figure(figsize=(10, 5))

shape = (2,6)
ax1 = plt.subplot2grid(shape, (0,0), colspan=2)
ax2 = plt.subplot2grid(shape, (0,2), colspan=2)
ax3 = plt.subplot2grid(shape, (0,4), colspan=2)
ax4 = plt.subplot2grid(shape, (1,1), colspan=2)
ax5 = plt.subplot2grid(shape, (1,3), colspan=2)

axarr = np.array([ax1, ax2, ax3, ax4, ax5])

i = 0
for beh in plot_data.keys():
    ax = axarr[i]
    ax.plot(range(len(plot_data[beh])), plot_data[beh][:, 0], label='x-axis')
    ax.plot(range(len(plot_data[beh])), plot_data[beh][:, 1], label='y-axis')
    ax.plot(range(len(plot_data[beh])), plot_data[beh][:, 2], label='z-axis')
    ax.set_ylim([-30, 30])
    ax.xaxis.set_major_locator(plt.MaxNLocator(3))
    ax.yaxis.set_major_locator(plt.MaxNLocator(3))
    ax.set_title(beh)
    i += 1
ax.legend(loc=10, bbox_to_anchor=(1.3, 0.2))

f.tight_layout(w_pad=-4.5)

plt.show()
```
